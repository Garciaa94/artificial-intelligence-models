```python
# paquete de computación científico y gestión de arreglos
import numpy as np
# Paquete para dividir datos (entrenamiento y testing)
from sklearn.model_selection import train_test_split
# paquete para plotear 
import matplotlib.pyplot as plt
# paquete uso de estructuras de datos y análisis de datos
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import linear_model
#paquete para crear interfaz gráfica
from tkinter import *
import tkinter as tk
from tkinter import messagebox
import easygui as eg
#paquete para abrir explorador de archivos
from tkinter.filedialog import askopenfilename
#paquete para exportar el modelo

```

```python
# Variables globales
X_train = 0
X_test = 0
y_train = 0
y_test = 0
X = 0
y = 0
regr = 0

def OpenFile(): # Función para cargar dataset
    try:
        Tk().withdraw() # we don't want a full GUI, so keep the root window from appearing
        filename = askopenfilename() # show an "Open" dialog box and return the path to the selected file
        #print(filename)
        dataset = pd.read_csv(filename) # Lectura del dataset
        filtered_data = dataset[(dataset['Word count'] <= 35000) & (dataset['# Shares'] <= 80000)]
        # ASignamos nuestra variable de entrada X para entrenamiento y las etiquetas Y.
        global X
        global y
        X = filtered_data[["Word count"]]
        y = filtered_data['# Shares']
        #Mensaje de confirmación 
        messagebox.showinfo(message="Dataset cargado correctamente", title="Mensaje")
    except:
        #Mensaje de error
        messagebox.showerror(message="No se eligió el archivo", title="Mensaje")

def Clean(): # Función para limpiar 
    train.set("")
    testing.set("")

def Train(): # Función entrenamiento
    try:
        global X_train
        global y_train
        global X_test
        global y_test
        global regr
        ts=(float(testing.get()))/100 # Porcentaje testing
        train.set(100-(int(testing.get()))) # Porcentaje de entrenamiento
        X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=ts)
        X_train = np.array(X)
        #y_train = filtered_data['# Shares'].values
        y_train = np.array(y)
        # Creamos el objeto de Regresión Linear
        regr = linear_model.LinearRegression()
        # Entrenamos nuestro modelo
        regr.fit(X_train, y_train)
        #Mensaje de información
        messagebox.showinfo(message="Modelo entrenado correctamente", title="Mensaje")
    except:
        #Mensaje de error
        messagebox.showerror(message="Ocurrió un error", title="Mensaje")
    

def GenerarModelo():  # Función para generar modelo entrenado
    try:
        #joblib.dump(regr,'modelo_regresion_articulo.pkl') # Nombre del modelo a generar
        if not regr:
            #Mensaje de información
            messagebox.showinfo(message="Modelo vacío", title="Mensaje")
        else:
            joblib.dump(regr,(eg.filesavebox(msg="Guardar modelo",
                         title="Machine Learning: ",
                         default='.pkl')))
            #Mensaje de información
            messagebox.showinfo(message="Modelo guardado correctamente", title="Mensaje") 
    except:
        #Mensaje de error
        messagebox.showerror(message="Ocurrió un error", title="Mensaje")
    


# Ventana (INTERFAZ)
ventana = Tk()
ventana.title("Machine Learning: Linear Regression") # Título del ventana

ventana = Frame(height=300,width=410) # El ancho y la altura de la interfaz
ventana.pack(padx=20,pady=20) # Relleno de la ventana

# ---------------------- LABEL ---------------------------
Label(ventana, justify="center",text="Entrenar el modelo para predecir el número de veces que \nun artículo será compartido en redes sociales",font=("Verdana",10)).place(x=0,y=0)

Label(text="Dataset Artículos: ",font=("Verdana",12)).place(x=0,y=85)
Label(text="% Train: ",  font=("Verdana",12)).place(x=0,y=120)
Label(text="% Testing: ",  font=("Verdana",12)).place(x=0,y=150)

# ---------------------- ENTRY ----------------------------

train= StringVar()
Entry(ventana, justify="center", textvariable=train, state="disable", font=("Arial",12)).place(x=150,y=100)

testing = StringVar()
Entry(ventana, justify="center", textvariable=testing, font=("Arial",12)).place(x=150,y=130)

# -------------------- BUTTON -----------------------------
Button(ventana, command=OpenFile, text="Cargar Archivo", font=("Verdana",10), background="gray").place(x=150, y=65)

Button(ventana, command=Train, text="Entrenar", font=("Verdana",10), background="aqua").place(x=55, y=190)
Button(ventana, command=GenerarModelo, text="Generar Modelo", font=("Verdana",10), background="aqua").place(x=126, y=190)
Button(ventana, command=Clean, text="Limpiar Datos", font=("Verdana",10), background="aqua").place(x=245, y=190)

ventana.mainloop() #Fin del programa
```
