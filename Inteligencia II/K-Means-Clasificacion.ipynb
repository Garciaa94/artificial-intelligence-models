```python
from sklearn.cluster import KMeans
from sklearn import datasets
from sklearn import metrics

from sklearn.datasets import load_iris
# Imports necesarios
import numpy as np
import pandas as pd
iris=load_iris()
print(iris)
```

```python
x=iris.data  # x de entrenamiento
x
#y=iris.target
```

```python
#Creacion de modelo de 2 a max 1000 interacciones
kmeans = KMeans(n_clusters=3,max_iter=1000) # si no se pone el maxm de iteraccuines por defecto 300 y si no se define el n-clster por defecto crea 8
```

```python
kmeans.fit(x)
```

```python
predicciones = kmeans.predict(x)
```

```python
predicciones
```

```python
y = iris.target
score = metrics.adjusted_rand_score(y,predicciones)
```

```python
T =kmeans.cluster_centers_
T
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')
```

```python
labels = kmeans.predict(x)
# Getting the cluster centers

colores=['red','green','blue']
asignar=[]
for row in labels:
    asignar.append(colores[row])

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(x[:, 0], x[:, 1], x[:, 2], c=asignar,s=60)
ax.scatter(T[:, 0], T[:, 1], T[:, 2], marker='*', c=colores, s=1000)
```
